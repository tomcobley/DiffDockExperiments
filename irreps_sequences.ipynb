```python
from e3nn import o3
```

```python
ns = 8
nv = 2
irrep_seq = [
                f'{ns}x0e',
                f'{ns}x0e + {nv}x1o',
                f'{ns}x0e + {nv}x1o + {nv}x1e',
                f'{ns}x0e + {nv}x1o + {nv}x1e + {ns}x0o'
            ]
```

```python
o3.Irreps("0e+1e+2e")
```

```python
tp = o3.FullTensorProduct( #o3.FullyConnectedTensorProduct(
    irreps_in1=irrep_seq[2],
    irreps_in2=o3.Irreps.spherical_harmonics(lmax=2),
)
print(tp)
tp.visualize()
```

```python
tp = o3.FullyConnectedTensorProduct(
    irreps_in1=irrep_seq[2],
    irreps_in2=o3.Irreps.spherical_harmonics(lmax=2),
    irreps_out=irrep_seq[3]
)
print(tp)
tp.visualize()
```
